# 第二步：准备一个函数可以直接生成一个列表

```python
# 准备一个行数运行
def create_name_list(file):

    nameoflist = []

    for line in file:
        line_s = line.strip()  # 字符中的多余的空格去掉
        line_s_s = line_s.split('         ')  #  分割成一个列表
        name_str = line_s_s[0]
        nameoflist.append(name_str)

    i = 1
    while i <= 3:
        nameoflist.pop(0)
        i += 1

    return nameoflist
```

# 第四步：准备函数的主体部分

```python
def create_data_list(file):
    dataoflist = []

    for line in file:
        if line.find('Band 1') != -1 and line.find('0.0') == -1:
            line_s = line.strip()
            dataoflist.append(line_s)
        elif line.find('Band 2') != -1 and line.find('0.0') == -1:
            line_s = line.strip()
            dataoflist.append(line_s)
        elif line.find('Band 3') != -1 and line.find('0.0') == -1:
            line_s = line.strip()
            dataoflist.append(line_s)
        elif line.find('Band 4') != -1 and line.find('0.0') == -1:
            line_s = line.strip()
            dataoflist.append(line_s)
        elif line.find('Band 5') != -1 and line.find('0.0') == -1:
            line_s = line.strip()
            dataoflist.append(line_s)
        else:
            pass

    return dataoflist

```

```python
# 拆分多个列表
def convert_multilist(list1, list2):

    i = 0
    a = 0
    list3 = []

    for item in list1 :
        attr2 = list2[a:a+5]
        list3.append(attr2)
        i += 5
        a = a + i

    return  list3
```

# 准备一个入口函数

```python
def datalist(file1,file2):
    list1 = create_name_list(file1)
    list2 = create_data_list(file2)
    list3 = convert_multilist(list1, list2)
    return list3


data_list = datalist(file1,file2)
```

# 数据处理成一个数据一个的list表

```python
def convert_data(list):
    data_list = []
    for sub_list in list:
        list_temp_1 = []

        for item in sub_list:
            str_temp = item[6:].strip()
            list_temp_1.append(str_temp)
            # print(listtemp_1)
        list_temp_2 = '  '.join(list_temp_1)
        list_temp_3 = list_temp_2.split('  ')
        # print(list_temp_3)
        list_temp_4 = []

        for item in list_temp_3:
            new_str = item.strip()
            list_temp_4.append(new_str)
            # print(list_temp_4)

        data_list.append(list_temp_4)
    return  data_list

convert_data(data_list)
```
